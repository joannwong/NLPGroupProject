## Pre-Trained Model

NER using BERT from the Transformers library by Hugging Face

```python
!pip install transformers
```

```python
from transformers import AutoModelForTokenClassification
from transformers import AutoTokenizer
from transformers import pipeline

model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")
tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer)
```

```python
text = "John Green is the author of Turtles All The Way Down."

ner_results = ner_pipeline(text)
print(ner_results)

for res in ner_results:
    print(res)
```

## Zero-shot Learning

```python
# Sample text
sample_text = "John Doe's email is johndoe@example.com, and his phone number is 555-1234."

# Run NER pipeline on the sample text
entities = ner_pipeline(sample_text)

# Print detected entities
for entity in entities:
    print(entity)
```
